```python
#Distribution of Diamonds by Diamond Cuts

sns.set_palette("Set2")
shapes = ['Round', 'Princess', 'Oval', 'Marquise', 'Pear', 'Cushion', 'Emerald', 'Asscher', 'Radiant', 'Heart', 'Cushion Modified']
plt.figure(figsize=(10, 6))
sns.catplot(x='cut', data=df, kind='count', order=shapes, aspect=2)
plt.title('Distribution of Diamonds by Diamond Cuts')
plt.xlabel('Cut')
plt.ylabel('Count')
plt.show()
```

```python
#Percentage Distribution of Diamonds by Diamond Cuts

shapes = ['Round', 'Princess', 'Oval', 'Marquise', 'Pear', 'Cushion', 'Emerald', 'Asscher', 'Radiant', 'Heart', 'Cushion Modified']
cut_counts = {cut: df['cut'].value_counts().get(cut, 0) for cut in shapes}
total_diamonds = sum(cut_counts.values())
for cut in shapes:
    percentage = (cut_counts[cut] / total_diamonds) * 100
    print(f"{cut} diamonds comprise {round(percentage, 1)}% of all diamonds in the set")
```

```python
#Distribution of Diamonds by Colors
sns.set_palette(palette="YlOrRd", n_colors=11)
colors = ['D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'unknown']
c = sns.catplot(x='color', data=df, kind='count', order=colors, aspect=2.5)
c.set(title="Distribution of Diamonds by Colors")
```

From the color of the diamonds we can observe that a high number of the diamonds are of a high - medium quality.


```python
#Percentage Distribution of Diamonds by Colors
colors = ['D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'unknown']
color_counts = {color: df['color'].value_counts().get(color, 0) for color in colors}
total_colored_diamonds = sum(color_counts.values())
for color in colors:
    percentage = (color_counts[color] / total_colored_diamonds) * 100
    print(f"{color} diamonds comprise {round(percentage, 1)}% of all colored diamonds in the set")
```

```python
#Distribution of Diamonds by Clarity
sns.set_palette(palette="YlOrRd", n_colors=14)
clarities = ['FL', 'IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'SI3', 'I1', 'I2', 'I3']
c = sns.catplot(x='clarity', data=df, kind='count', order=clarities, aspect=2.5)
c.set(title='Distribution of Diamonds by Clarity')
```

We can observe that the majority of the diamonds are in between the Very, Very Slightly Included (VVS1) and Slightly Included(Sl1).

```python
#Percentage Distribution of Diamonds by Clarity
clarities = ['FL', 'IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'SI3', 'I1', 'I2', 'I3']
clarity_counts = {clarity: df['clarity'].value_counts().get(clarity, 0) for clarity in clarities}
total_diamonds = sum(clarity_counts.values())
for clarity in clarities:
    percentage = (clarity_counts[clarity] / total_diamonds) * 100
    print(f"{clarity} diamonds comprise {round(percentage, 1)}% of all diamonds in the set")
```

```python
#Correlation between Color and Clarity
clarity_order = ['FL', 'IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'SI3', 'I1', 'I2', 'I3']
color_order = ['D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'unknown']
plt.figure(figsize=(10, 6))
heatmap_data = df.groupby(['clarity', 'color']).size().unstack().reindex(index=clarity_order, columns=color_order)

sns.heatmap(heatmap_data, annot=True, fmt='0.0f', cmap='coolwarm', cbar=False)
plt.title('Correlation between Color and Clarity')
plt.xlabel('Color')
plt.ylabel('Clarity')

plt.show()

```

From the heatmap we can observe a correlation between color and clarity. As the color of the diamonds becomes less and less colorless, the clarity gradually decreases.

```python
#Diamond Count by Carat Weight
carat_counts = df['carat_weight'].value_counts().reset_index()
carat_counts.columns = ['carat_weight', 'count']
carat_counts = carat_counts.sort_values(by='carat_weight')
plt.figure(figsize=(12, 6))
plt.bar(carat_counts['carat_weight'], carat_counts['count'], color='orange')
plt.title('Diamond Count by Carat Weight')
plt.xlabel('Carat Weight')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.xlim(0, 4)
plt.show()
```

There is a negative correlation between Carat Weight and the Count of the diamonds. The number of diamonds corresponding to a Carat Weight decreases as the Carat Weight increases.

```python
#Scatter Plot of Carat Weight vs Total Sales Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat_weight', y='total_sales_price', data=df, alpha=0.5, color ="green")
plt.title('Scatter Plot: Carat Weight vs. Total Sales Price')
plt.xlabel('Carat Weight')
plt.ylabel('Total Sales Price')
plt.show()

```

There is a positive correlation between Carat Weight and Total Sales Price. As the Carat Weight increases, the Total Sales Price increases as well.

```python
#Box plot of Carat Weight vs Cut
plt.figure(figsize=(16, 8))
plt.subplot(131)
sns.boxplot(x='cut', y='carat_weight', data=df)
plt.title('Carat Weight vs. Cut')
plt.xticks(rotation=55)
#Box plot of Carat Weight vs Color
plt.subplot(132)
sns.boxplot(x='color', y='carat_weight', data=df)
plt.title('Carat Weight vs. Color')
plt.xticks(rotation=55)
#Box plot of Carat Weight vs Clarity
plt.subplot(133)
sns.boxplot(x='clarity', y='carat_weight', data=df)
plt.title('Carat Weight vs. Clarity')
plt.xticks(rotation=55)

plt.tight_layout()
plt.show()

```

```python
categorical_columns = [feature for feature in df.columns if df[feature].dtypes =='O']
numerical_columns=[feature for feature in df.columns if df[feature].dtypes !='O']
for cat in categorical_columns:
    sns.barplot(df,y='total_sales_price',x=cat)
    plt.xticks(rotation=75)
    plt.title(cat)

    plt.show()

for cat in numerical_columns:
    sns.relplot(df,y='total_sales_price',x=cat, hue= 'cut_quality' )
    plt.xticks(rotation=75)
    plt.title(cat)

    plt.show()
```
